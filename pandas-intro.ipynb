#### 14. Make a new DataFrame containing just the crime categories and the number of crimes per category.

```python
df_sf_crime.head()
```

```python
categories_list = list(df_sf_crime['category'].unique())
sorted_categories_list = sorted(categories_list)

crimes_per_category = list(df_sf_crime.groupby('category')['category'].size())
d = {'category': sorted_categories_list,'crimes': crimes_per_category}
new_Data_Frame = pd.DataFrame(d)
new_Data_Frame
# print categories_list
# print crimes_per_category
# print type(new_Data_Frame)

# df = df_sf_crime.groupby('category')['category'].size()
# df
# df_sf_crime['category'].value_counts()

```

```python
df_category = df_sf_crime['category'].value_counts().to_frame().reset_index().head()
df_category.columns = ['category', 'number_of_crimes']
df_category
```
